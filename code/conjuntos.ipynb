```python
import matplotlib.pyplot as plt


efectividad = [12, 17, 25, 25]
altura = [170, 195, 220, 220]
sprint = [0, 0, 10, 15]
y = [0, 1, 1, 0]
```

```python
plt.figure(figsize = (10, 5))
plt.plot(efectividad, y)
plt.title('Función de membresía de la característica "Buen anotador"')
plt.xlabel('Número de tiros')
plt.ylabel('Grado de pertenencia')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(altura, y)
plt.title('Función de membresía de la característica "Alto"')
plt.xlabel('Altura (cm)')
plt.ylabel('Grado de pertenencia')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(sprint, y)
plt.title('Función de membresía de la característica "Rápido"')
plt.xlabel('Tiempo de sprint (s)')
plt.ylabel('Grado de pertenencia')
plt.show()
```
